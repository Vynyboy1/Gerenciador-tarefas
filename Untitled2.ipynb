```python
import numpy as np
import pandas as pd
```

```python
#leitor do arquivo
df = pd.read_csv("https://pycourse.s3.amazonaws.com/bike-sharing.csv")
```

```python
df.head()
```

```python
print(f'a media da coluna windspeed é {df["windspeed"].mean()}')
print(f'a media da coluna temp é {df["temp"].mean()}')

```

```python
#total de bicicletas registradas por anos(0-2011, 1 - 2012)
df[['year','registered','total_count']].groupby(by= 'year').sum()
```

```python
df[['season','total_count']].groupby(by= 'season').mean().sort_values('total_count')
```

```python
df[['hour','total_count']].groupby(by= 'hour').mean().sort_values('total_count')
```

```python
df[['season','total_count']].groupby(by= 'season').sum().sort_values('total_count')
```

```python
df[['weekday','total_count']].groupby(by= 'weekday').mean().sort_values('total_count')
```

```python
df.groupby(by='weekday')
```

```python
df2 = df[df['weekday']==3]
df2[['hour','total_count','registered','weekday']].groupby(by='hour').mean().sort_values('total_count')
```
